```python
# Farmers Survey

# Specify the file path
file_path = r'C:\Users\chimi\Desktop\Python Data Science Projects\Cocoa Production\Farmers Survey\farmers survey.xlsx'

# Read the Excel file
farmers_survey = pd.read_excel(file_path)

# Print the first few rows of the DataFrame
print(farmers_survey.head())
```

```python
farmers_survey.drop(columns=['Stand properties','Soil nutrients'],inplace=True)
```

```python
# Create a new variable 'Shade_Tree.Density (shade trees ha-1)' by multiplying 'shade trees density ha' and 'Plot size ha'

farmers_survey['Shade_Tree.Density (shade trees ha-1)'] = farmers_survey['shade trees density ha'] * farmers_survey['Plot size ha']
```

```python
# Drop the column
farmers_survey.drop(columns=['shade trees density ha'],inplace=True)
```

```python
# Convert Cocoa Density from per acre to 1 acre
farmers_survey['Cocoa_density (cocoa trees ha-1)'] = farmers_survey['Cocoa plant density'] / farmers_survey['Plot size ha']
```

```python
# Drop cocoa plants
farmers_survey.drop(columns=['Cocoa plant density'],inplace=True)
```

```python
# Convert Shade tree species.ha to Shade tree species ha-1
farmers_survey['Shade tree species ha-1'] = farmers_survey['Shade tree species.ha'] / farmers_survey['Plot size ha']
```

```python
# Drop the column
farmers_survey.drop(columns=['Shade tree species.ha'],inplace=True)
```

```python
# Convert Carbon.% to Carbon (t ha-1)
farmers_survey['Carbon (t ha-1)'] = farmers_survey['Carbon.%'] / 100 * (farmers_survey['Plot size ha'] * 10000)

# Convert Total N% to Nitrogen (t ha-1)
farmers_survey['Nitrogen (t ha-1)'] = farmers_survey['Total N%'] / 100 * (farmers_survey['Plot size ha'] * 10000)

# Convert Soil organic matter % to Soil organic matter (t ha-1)
farmers_survey['Soil organic matter (t ha-1)'] = farmers_survey['Soil organic matter %'] / 100 * (farmers_survey['Plot size ha'] * 10000)

# Convert Ca.cmol.kg to Ca.ex (mg kg-1)
farmers_survey['Ca.ex (mg kg-1)'] = farmers_survey['Ca.cmol.kg'] * 10000.

# Convert Mg.cmol.kg to Mg.ex (mg kg-1)
farmers_survey['Mg.ex (mg kg-1)'] = farmers_survey['Mg.cmol.kg'] * 10000.

# Convert K.cmol.kg to K.ex (mg kg-1)
farmers_survey['K.ex (mg kg-1)'] = farmers_survey['K.cmol.kg'] * 10000.

# Convert Na.cmol.kg to Na.ex (mg kg-1)
farmers_survey['Na.ex (mg kg-1)'] = farmers_survey['Na.cmol.kg'] * 10000.

# Convert P.cmol.kg to Phosphorus (kg ha-1)
farmers_survey['Phosphorus (kg ha-1)'] = farmers_survey['P.cmol.kg'] * 10000.



# Convert Bulk.density (g cm-³) to Bulk.density (g cm-³)
# This column remains the same

# Drop the original columns
farmers_survey = farmers_survey.drop(columns=['Carbon.%', 'Total N%', 'Soil organic matter %', 'Ca.cmol.kg',
                                              'Mg.cmol.kg', 'K.cmol.kg', 'Na.cmol.kg', 'P.cmol.kg',
                                              'Exch. A (Al+H)/cmol/kg', 'E.C.E.C /cmol/kg'])
                                              
# Print the updated DataFrame
print(farmers_survey.head())
```

```python
farmers_survey.columns
```

```python
# Combine temperature columns
temperature_columns = ['Temp.Mean.Oct.2015 (°C)', 'Temp.Mean.Oct.2016 (°C)',
                       'Temp.Mean.Feb.2016 (°C)', 'Temp.Mean.Feb.2017 (°C)']
agroforest_survey['Avg_Temp'] = agroforest_survey[temperature_columns].mean(axis=1)

# Combine RH columns
rh_columns = ['RH.mean.Oct.2015 (%)', 'RH.mean.Oct.2016 (%)',
              'RH.mean.Feb.2016 (%)', 'RH.mean.Feb.2017 (%)']
agroforest_survey['Avg_RH'] = agroforest_survey[rh_columns].mean(axis=1)

# Combine soil moisture columns
soil_moisture_columns = ['Soil.moisture.Aug.2015 (%)', 'Soil.moisture.Feb.2016 (%)']
agroforest_survey['Avg_Soil_Moisture'] = agroforest_survey[soil_moisture_columns].mean(axis=1)

# Combine light columns
light_columns = ['Light_Sep.2016 (%)', 'Light_Feb.2016 (%)']
agroforest_survey['Avg_Light'] = agroforest_survey[light_columns].mean(axis=1)

# Drop the original columns
agroforest_survey = agroforest_survey.drop(columns=temperature_columns + rh_columns + soil_moisture_columns + light_columns)

# Print the updated DataFrame
print(agroforest_survey.head())
```

```python
# Max temperature, relative humidity, and  columns

temperature_columns = ['Temp.Max.Oct.2015 (°C)', 'Temp.Max.Oct.2016 (°C)',
                       'Temp.Max.Feb.2016 (°C)', 'Temp.Max.Feb.2017 (°C)']
agroforest_survey['Avg_Temp_Max'] = agroforest_survey[temperature_columns].mean(axis=1)

# Combine relative humidity columns
rh_columns = ['RH.Min.Oct.2015 (%)', 'RH.Min.Oct.2016 (%)',
              'RH.Min.Feb.2016 (%)', 'RH.Min.Feb.2017 (%)']
agroforest_survey['Avg_RH_Min'] = agroforest_survey[rh_columns].mean(axis=1)

# Drop the original columns
agroforest_survey.drop(columns=temperature_columns + rh_columns, inplace=True)

# Print the updated DataFrame
print(agroforest_survey.head())

```
