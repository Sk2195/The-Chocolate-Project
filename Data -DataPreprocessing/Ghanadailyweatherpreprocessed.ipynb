```python
import numpy as np
import pandas as pd
```

```python
# Historical Weather DataSet Pre-processing-EDA
# OR using raw string
weather_df = pd.read_csv(r"C:\Users\chimi\Desktop\Python Data Science Projects\Cocoa Production\Data -1\Ghana Weather dataset 2010-01-01 to 2023-06-03.csv")
```

```python
# Drop irrelevant columns filled with 0's and missing values
weather_df.drop(columns=['snow','snowdepth','windgust','severerisk','sunset','sunrise',
                     'severerisk','icon','stations','moonphase','sealevelpressure','description',
                    'conditions'],inplace=True)
```

```python
weather_df.columns
```

```python
# New shape
weather_df.shape
```

```python
# Convert the date to proper format
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'], format='%Y-%m-%d')
```

```python
weather_df.head(5)
```

```python
weather_df[['Country','Region']] = weather_df['name'].str.split(' ',expand=True)
```

```python
weather_df.drop(columns=['name'],inplace=True)
```

```python
weather_df.head(5)
```

```python
# Let's drop more irrelevant column to reduce the shape
weather_df.drop(columns=['feelslikemax','feelslikemin', 'feelslike','precipprob',
       'precipcover', 'preciptype'], inplace=True)
```

```python
weather_df.shape
```

```python
#Weather types
weather_df.dtypes.value_counts()
```

```python
# Filter in data from the past 10 years
mask = (weather_df['datetime'] >= '2013-01-01')
filtered_weather_df = weather_df.loc[mask]
```
