```python
import pandas as pd
import rasterio
# Just opening vrt file. Other files just contained metatags
# Used Golden Software to look over grd and gri files.
# They are meta tags only.
 
```

```python

# Open the VRT file
with rasterio.open(r"C:\Users\chimi\Desktop\Python Data Science Projects\The chocolate Project\Data Collection\Elevation-Data\GHA_alt.vrt") as src:
    # Read the raster data as a 2D numpy array
    raster_data = src.read(1)

    # Get the GeoTransform
    transform = src.transform

    # Get the number of rows and columns
    rows, cols = raster_data.shape

    # Initialize lists to store coordinates and elevation data
    coordinates = []
    elevation_data = []

    # Iterate over each pixel and calculate coordinates
    for row in range(rows):
        for col in range(cols):
            # Calculate the x and y coordinates
            x = transform[0] + col * transform[1] + row * transform[2]
            y = transform[3] + col * transform[4] + row * transform[5]

            # Append the coordinates and elevation value to the lists
            coordinates.append((x, y))
            elevation_data.append(raster_data[row, col])

# Create a DataFrame from the coordinates and elevation data
data = {'Coordinate': coordinates, 'Elevation': elevation_data}
df = pd.DataFrame(data)

# Save the DataFrame as a CSV file
df.to_csv('elevation_data.csv', index=False)
```
