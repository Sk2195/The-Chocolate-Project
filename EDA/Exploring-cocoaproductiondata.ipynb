```python
# Select the columns for which you want to visualize histograms
columns_to_plot = ['Area harvested(ha)', 'Production(tonnes)', 'Yield(hg/ha)']

# Create individual histograms for each variable
for column in columns_to_plot:
    plt.figure(figsize=(12, 6))
    sns.histplot(data=cocoa_data, x=column, bins=10, kde=True)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.grid(True)
    plt.show()
```

```python
# What is the overall trend in cocoa production over the years?
# From the early 1980s to the early 1990s, cocoa production experienced a general downward trend, reaching its lowest 
# point around the early 1990s.
# Starting from the mid-1990s, there was a significant increase in cocoa production, with occasional fluctuations, 
# until reaching a peak in the early 2000s.
# After the peak in the early 2000s, cocoa production experienced a mixed pattern with some
# ups and downs, but without reaching the previous peak levels.
# 2010 onwards,cocoa production appears to be reletively stable,fluctuations, but not showing a clear increasing or decreasing trend.


plt.figure(figsize=(10, 6))
plt.plot(cocoa_data['Year'], cocoa_data['Production(tonnes)'], marker='o')
plt.xlabel('Year')
plt.ylabel('Total Production (tonnes)')
plt.title('Trend in Cocoa Production over the Years')
plt.grid(True)
plt.show()
```

```python
#What is the distribution of yields (in hg/ha) across different decade?
# Box plot indicates that the 2010s has the highest median yields
# Are there any significant variations or outliers? 
# Outliers indicate values that deviate from the overall pattern 
#, but are not substantially different or influential compared to the majority of the data
# Convert 'Year' column to datetime
cocoa_data['Year'] = pd.to_datetime(cocoa_data['Year'], format='%Y')

# Create a new column 'Decade' based on the year
cocoa_data['Decade'] = cocoa_data['Year'].dt.year // 10 * 10

# Create a box plot of yields for each decade
plt.figure(figsize=(10, 6))
sns.boxplot(data=cocoa_data, x='Decade', y='Yield(hg/ha)')
plt.xlabel('Decade')
plt.ylabel('Yield (hg/ha)')
plt.title('Distribution of Yields across Decades')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
#How has the area harvested for cocoa production changed over the years? 
# Are there any noticeable trends or patterns?
# In the early 2000s, the area harvested for cocoa production reached its peak.
#Since 2010, there has been consistent fluctuation in the area harvested, with either a decreasing or increasing trend.
# Between 1980 and 1990, the area harvested for cocoa production hit an all-time low.

# Sort the data by year
sorted_data = cocoa_data.sort_values('Year')

# Plot the trend of area harvested over time
plt.figure(figsize=(10, 6))
plt.plot(sorted_data['Year'], sorted_data['Area harvested(ha)'])
plt.xlabel('Year')
plt.ylabel('Area Harvested (ha)')
plt.title('Trend in Area Harvested for Cocoa Production')
plt.grid(True)
plt.show()
```

```python
# Is there corrleation between area harvested and cocoa production?
# There is a moderately strong positive correlation between these two variables. 
# It suggests that there is a tendency for an increase in the area harvested to be associated with higher cocoa production.


# Plot the correlation between area harvested and cocoa production
plt.figure(figsize=(10, 6))
plt.scatter(cocoa_data['Area harvested(ha)'], cocoa_data['Production(tonnes)'])
plt.xlabel('Area Harvested (ha)')
plt.ylabel('Cocoa Production (tonnes)')
plt.title('Correlation between Area Harvested and Cocoa Production')
plt.grid(True)
plt.show()
```

```python
# Is there corrleation between area harvested and cocoa yield?
# There is weak positive correleation between these two variables.
# It suggests that there is a slight increase in the area harvested to be associated with the higher cocoa yield.
# Plot the correlation between area harvested and cocoa yield
plt.figure(figsize=(10, 6))
plt.scatter(cocoa_data['Area harvested(ha)'], cocoa_data['Yield(hg/ha)'])
plt.xlabel('Area Harvested (ha)')
plt.ylabel('Cocoa Yield (hg/ha)')
plt.title('Correlation between Area Harvested and Cocoa Yield')
plt.grid(True)
plt.show()
```
