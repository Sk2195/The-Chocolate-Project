```python
# Aggregrating the soil dataset for regional analysis
import numpy as np
import pandas as pd
import os
```

```python
def read_files(directory):
    file_list = os.listdir(directory)
    dataframes = {}

    for idx, file in enumerate(file_list):
        if file.endswith('.csv') or file.endswith('.xlsx') or file.endswith('.xls'):
            file_path = os.path.join(directory, file)
            df_name = 'df{}'.format(idx)
            if file.endswith('.csv'):
                df = pd.read_csv(file_path, encoding='utf-8')
            else:
                df = pd.read_excel(file_path)
            dataframes[df_name] = df

    return dataframes

# Specify the directory containing the CSV and Excel files
directory = r'C:\Users\chimi\Desktop\Python Data Science Projects\Cocoa Production\Farmers Survey'

# Call the function to read the files
dataframes = read_files(directory)

# Access the individual DataFrames using their respective variable names
for df_name, df in dataframes.items():
    exec('{} = df'.format(df_name))
    print(df_name)
    print(df.head())
    print()

```
